## 📥 Step 2: Data Loading
This step clones the dataset repository and imports the registered demographic records (1993/1997–2025) for Korea and Thailand.

### Data Source Details:
* **Repository:** [my-city-in-2040](https://github.com/PNU-THAI-STEM-COMPETITION-2026/my-city-in-2040)
* **Data Format:** Cleaned demographic CSV files, separated into Annual data and 5-Year snapshot files.
* **Reference Date:** December 31 observations through 2025. Partial 2026 monthly data are not included.
* **Important Thailand Coverage Note:** The processed Thailand totals include only records assignable to the standard age groups. DOPA special categories—including lunar-year birth records, central house registration, non-Thai nationals, and people in the process of moving in recent source files—are excluded because they do not have a compatible age-group breakdown. Therefore, totals from these CSV files are lower than DOPA's full registered-population totals.

```python
# In Google Colab, we need to clone the repository to get the data files.
# If running locally in the cloned repository, this clone step will be skipped.
if not os.path.exists("my-city-in-2040") and not os.path.exists("data"):
    print("Cloning Track B repository...")
    !git clone https://github.com/PNU-THAI-STEM-COMPETITION-2026/my-city-in-2040.git
else:
    print("Data repository folder found!")

# Determine the base directory
base_dir = "my-city-in-2040" if os.path.exists("my-city-in-2040") else "."

# ----------------------------------------------------
# Load ALL Demographics CSV Data files (Thailand & Korea)
# ----------------------------------------------------
th_annual_path = f"{base_dir}/data/THA/population_age_sex.csv"
kor_annual_path = f"{base_dir}/data/KOR/population_age_sex.csv"
th_5year_path = f"{base_dir}/data/THA/population_age_sex_5year.csv"
kor_5year_path = f"{base_dir}/data/KOR/population_age_sex_5year.csv"

# 1. Load and combine Annual population data
annual_dfs = []
if os.path.exists(th_annual_path):
    df_th_annual = pd.read_csv(th_annual_path)
    df_th_annual["Country"] = "Thailand"
    annual_dfs.append(df_th_annual)
if os.path.exists(kor_annual_path):
    df_kor_annual = pd.read_csv(kor_annual_path)
    df_kor_annual["Country"] = "South Korea"
    annual_dfs.append(df_kor_annual)

if annual_dfs:
    df_population_annual = pd.concat(annual_dfs, ignore_index=True)
    print("\n--- Annual Population Data Loaded & Merged ---")
    print(f"Shape: {df_population_annual.shape}")
    print(f"Countries: {df_population_annual['Country'].unique()}")
    display(df_population_annual.head())
else:
    print("\nNo annual population files found.")

# 2. Load and combine 5-Year snapshot data
five_year_dfs = []
if os.path.exists(th_5year_path):
    df_th_5year = pd.read_csv(th_5year_path)
    df_th_5year["Country"] = "Thailand"
    five_year_dfs.append(df_th_5year)
if os.path.exists(kor_5year_path):
    df_kor_5year = pd.read_csv(kor_5year_path)
    df_kor_5year["Country"] = "South Korea"
    five_year_dfs.append(df_kor_5year)

if five_year_dfs:
    df_population_5year = pd.concat(five_year_dfs, ignore_index=True)
    print("\n--- 5-Year Snapshot Population Data Loaded & Merged ---")
    print(f"Shape: {df_population_5year.shape}")
    display(df_population_5year.head())
else:
    print("\nNo 5-year population files found.")
```
